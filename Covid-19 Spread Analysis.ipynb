```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/india_covid_19_state_wise_info.csv")
```

```python
c=df.iloc[:37,2].values
r=df.iloc[:37,3].values
d=df.iloc[:37,4].values
a=df.iloc[:37,5].values
x=list(df.iloc[:37,0])
 
plt.figure(figsize=(25,10))
ax=plt.axes()
 
ax.set_facecolor('gray')
ax.grid(linewidth=0.4, color='white')
 
 
plt.xticks(rotation='vertical',
           size='20',
           color='white')
 
plt.yticks(size='20',color='white')
 
 
ax.set_xlabel('\nStates',size=25,
              color='green')
ax.set_ylabel('No. of cases\n',size=25,
              color='green')
 
 
plt.tick_params(size=20,color='white')
 
 
ax.set_title('State wise cases in Inida\n',
             size=50,color='blue')
 
plt.bar(x,c,label='re')
plt.bar(x,r,label='re',color='green')
plt.bar(x,d,label='re',color='red')
plt.bar(x,a,label='re',color='yellow')
 
for i,j in zip(x,c):
    ax.annotate(str(int(j)),
                xy=(i,j+4),
                color='white',
                size='15')
 
plt.legend(['Confirmed','Recovered','Deaths','Active'],
           fontsize=20)
```
